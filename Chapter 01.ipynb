# **Introduction to Machine Learning**

## What Is Machine Learning?

> "Machine learning is the field of study that gives computers the *ability to learn* without being explicitly programmed."  
> – Arthur Samuel, 1959

> "Machine learning (ML) is a field of study in artificial intelligence focused on the development and analysis of statistical algorithms that can learn from data and generalize to unseen data, and thus *perform tasks without explicit instructions*."  
> – Wikipedia, 2025

**Example**: Email spam filter.

---

### Key Terms

- **Training Set**: The set of examples used by machine learning algorithms during the training process.  
- **Training Instance (or Sample)**: An individual example in the training set.  
- **Model**: The component of a machine learning system that learns from data and makes predictions (e.g., Neural Networks, Random Forest).  
- **Accuracy**: A performance measure indicating the ratio of correct outputs produced by a machine learning system.


## Why Use Machine Learning?

Machine learning offers several advantages over traditional programming approaches:

1. **Simpler and More Efficient Programs**  
   Machine learning-based programs are often shorter, easier to maintain, and more accurate compared to traditional ones. For instance, a traditional spam filter would require frequent updates and rely on a complex set of predefined rules to identify spam. In contrast, machine learning can automate this process by learning patterns from data.

2. **Solving Problems Without Traditional Solutions**  
   Machine learning excels at tackling problems that lack well-defined traditional solutions, such as speech recognition or image processing.

3. **Uncovering Hidden Patterns**  
   Machine learning identifies previously undetected correlations and behaviors within data, enabling a deeper understanding of the problem at hand.

Machine learning is particularly effective in the following scenarios:
- Traditional problems requiring numerous solutions and extensive fine-tuning.
- Complex problems where no traditional solutions exist.
- Dynamic and rapidly changing environments.
- Analyzing large datasets to extract valuable and complex insights.

---

**Data Mining**: Machine learning is instrumental in processing vast amounts of data to uncover hidden patterns and insights—a process known as data mining.

## Types of Machine Learning Systems

### Training Supervision
---
Machine learning systems can be classified based on the amount and type of supervision they receive during the training phase. These categories can be further divided into subcategories:

#### 1. Supervised Learning

In supervised learning, the training dataset includes the desired outcomes, known as *labels*. This approach is suitable for tasks such as:  
- **Classification**: Assigning categories, e.g., a spam filter classifies emails as "spam" or "not spam."  
- **Regression**: Predicting a continuous value, such as estimating a car's price based on its *features*.  

**Key Notes:**
- Some regression models, like logistic regression, can also handle classification problems by predicting the probability of a given outcome (e.g., the likelihood an email is spam).
- The terms *target* and *label* are often associated with supervised learning:
  - *Target*: Commonly used in regression tasks.
  - *Label*: Typically refers to classification tasks.  
- *Features* are also referred to as *predictors* or *attributes* in certain contexts.

#### 2. Unsupervised Learning

In unsupervised learning, the training data is unlabeled. These algorithms aim to discover the underlying structure in the data. Common tasks include:  
- **Clustering**: Grouping data points, e.g., hierarchical clustering.  
- **Dimensionality Reduction**: Simplifying data while retaining important information. For example:
  - Combining highly correlated features into a single feature (known as *feature extraction*).  

**Other Tasks:**
- **Anomaly Detection**: Identifying outliers, such as unusual credit card transactions or manufacturing defects.  
- **Novelty Detection**: Detecting new patterns in extremely clean datasets that exclude instances the system is trained to identify.  
- **Association Rule Learning**: Extracting relationships between features within large datasets.

#### 3. Semi-Supervised Learning

Semi-supervised algorithms work with partially labeled data. A common example is Google Photos, where some images are labeled, and the system learns to generalize. Most semi-supervised systems combine supervised and unsupervised learning techniques.

#### 4. Self-Supervised Learning

Self-supervised learning involves generating a fully labeled dataset from an unlabeled one. For example:  
- Masking parts of an image and using the original image as the desired output. This allows the model to learn patterns and structure in the data.  
- Fine-tuning the model afterward enables it to classify images based on user-defined categories.  

**Transfer Learning**: Reusing knowledge from one learning task to solve another is known as *transfer learning*. It is especially important in *deep neural networks*.

#### 5. Reinforcement Learning

In reinforcement learning, the system, referred to as an *agent*, interacts with its environment by performing tasks and receiving rewards or penalties.  
- The agent learns a *policy*, which determines the optimal actions to maximize cumulative rewards in the given environment.

**Applications**: Robotics, game AI, and autonomous systems.

### Batch vs. Online Learning
---

This distinction is based on whether the system can learn incrementally from a continuous stream of incoming data.

#### Batch Learning

In **batch learning**, the system cannot learn incrementally. Instead, it must be trained using the entire available dataset at once. After training, the model is deployed to production and applies what it has learned to make predictions. This approach is also known as **offline learning**.

**Key Characteristics:**
- **Static Training**: The model is trained once and remains static until retrained.  
- **Performance Decay**: Over time, as new data becomes available and the environment changes, the model's performance may degrade. This phenomenon is known as **data drift** or **model rot**.  
- **Retraining Required**: To maintain performance, the model needs to be retrained periodically with updated data. However, this process can be computationally expensive and impractical for rapidly changing datasets.  

**Limitations:**  
Batch learning is not suitable for scenarios where:
- Data changes rapidly.
- The system needs to adapt autonomously in real-time.

#### Online Learning

In **online learning**, the model is trained incrementally by feeding data in small **mini-batches** or even one instance at a time. This method is fast, cost-effective, and well-suited for:
- **Rapidly Changing Datasets**: Adapting to new data quickly.  
- **Large Datasets**: Handling datasets too large to fit into memory at once, a process known as **out-of-core learning**.  

**Note:** Despite the term *online learning*, the training may still occur offline, which can be a source of confusion.

**Key Parameter:**  
- **Learning Rate**: Determines how quickly the model adapts to new data:
  - A **high learning rate** allows the model to learn quickly but may result in forgetting past data (known as catastrophic forgetting).
  - A **low learning rate** ensures gradual learning and reduces sensitivity to noise in the data.

**Challenges:**  
- **Data Quality**: Poor-quality data can cause the model's performance to deteriorate rapidly.  
- **Monitoring Required**: Continuous monitoring is essential to ensure the model’s performance remains optimal. Implementing automated **anomaly detection algorithms** can help identify and address issues with incoming data streams.

### Instance-Based vs. Model-Based Learning
---

This classification is based on how models generalize from data.

#### Instance-Based Learning

In **instance-based learning**, the system memorizes examples and uses a **similarity measure** to classify new inputs. Instead of creating a general model, it compares new instances directly to stored examples.

**Key Characteristics:**
- Relies on the similarity between new data points and existing examples.  
- Does not explicitly generalize; instead, it "learns by heart."  
- Examples: **k-Nearest Neighbors (k-NN)** and **Case-Based Reasoning (CBR)**.

#### Model-Based Learning

In **model-based learning**, the system uses a **model** to make predictions. The process involves discovering relationships between input features and output values by fitting the data to a predefined model.

**Example:**  
Consider the relationship between Life Satisfaction and GDP. A clear trend might emerge, suggesting the use of a **linear model** to represent the data.  

**Steps in Model-Based Learning:**
1. **Model Selection**: Choose a model that best captures the relationship (e.g., linear, polynomial, or other models).  
2. **Parameter Optimization**: Define the parameters (e.g., $( \theta_0, \theta_1 )$) to minimize prediction errors.  

**Performance Metrics:**  
To evaluate and optimize the model, you need a **performance measure**, which can be:
- **Fitness Function (Utility Function)**: Measures how well the model performs.
- **Loss Function**: Measures how poorly the model performs (e.g., Mean Squared Error).

**Goal:**  
Optimize parameters by minimizing the loss function or maximizing the fitness function to ensure accurate predictions.

## Main Challenges of Machine Learning

The primary goal of machine learning is to select the correct model and train it on the right data. The main challenges stem from either a **bad model** or **bad data**.

---

### 1. Insufficient Quantity of Training Data

In a 2001 paper by Microsoft researchers Michele Banko and Eric Brill, it was shown that various machine learning algorithms performed almost identically well on complex tasks when provided with enough data. This highlights the trade-off between investing in algorithm development versus data collection. A larger dataset often yields better performance than more sophisticated algorithms with limited data.

---

### 2. Nonrepresentative Training Data

It is critical to use a training set that is representative of the real-world cases you aim to generalize. This can be challenging due to:
- **Sampling Noise**: Occurs when the dataset is too small, leading to randomness and lack of representativeness.
- **Sampling Bias**: Even with large datasets, if the sampling method is flawed, the training data will not accurately represent the target population.

---

### 3. Poor Quality Data

Training data with errors, noise, or outliers (e.g., due to measurement issues) makes it harder for the model to detect patterns and generalize effectively. Addressing poor-quality data involves:
- Deciding whether to discard or include outliers.
- Handling missing features or values appropriately.

---

### 4. Irrelevant Features

The performance of a machine learning model depends on the quality of the input features. Irrelevant or redundant features lead to poor results. **Feature engineering** is a crucial step and includes:
- **Feature Selection**: Choosing the most relevant features.
- **Feature Extraction**: Combining similar features to create new ones.

---

### 5. Overfitting the Training Data

**Overfitting** occurs when the model performs exceptionally well on the training data but fails to generalize to unseen data. This is often due to:
- Complex models (e.g., neural networks) capturing noise or spurious patterns in the data.
- Insufficient training data or excessive noise in the dataset.

**Solutions to Overfitting:**
- Simplify the model by reducing parameters.
- Gather more training data.
- Reduce noise in the training data.
- **Regularization**: Constrain the model to make it simpler and less prone to overfitting. The degree of regularization is controlled by **hyperparameters**.

**Hyperparameters**: Parameters of the learning algorithm, not the model itself. These are set before training and remain constant throughout. For example, if you set a high regularization parameter, the model may become too flat, avoiding overfitting but underperforming.

**Hyperparameter tuning** is a crucial part of machine learning to balance model complexity and generalization.

---

### 6. Underfitting the Training Data

**Underfitting** is the opposite of overfitting. It happens when the model is too simple to capture the underlying patterns in the data. This results in poor performance on both training and test data.

**Solutions to Underfitting:**
- Select a more complex model with additional parameters.
- Use better features.
- Reduce constraints or regularization applied to the model.

---

### 7. Testing and Validation

To evaluate your model, split your dataset into training and testing sets from the beginning. The **error rate** on the test data is referred to as the **generalization error** (or **out-of-sample error**). A low training error combined with a high generalization error indicates an **overfitting model**.

**Typical Split:**
- **80/20 Split**: Commonly, 80% of the data is used for training, and 20% is used for testing.  
- For **large datasets**, the test set can be smaller since it still provides robust evaluations.

---

### 8. Hyperparameter Tuning and Model Selection

When selecting models or optimizing hyperparameters, comparing models and training them with different hyperparameter values is standard practice. However, repeatedly testing models on the test set can make the model overly fit to the test set rather than generalizing well.

**Solutions:**
1. **Holdout Validation**:  
   - Split the training set into a **reduced training set** and a **validation set** (also called a **dev set**).  
   - Train multiple models on the reduced training set and evaluate them on the validation set.  
   - Select the best-performing model, retrain it on the **entire training set**, and test it on the test set for final evaluation.

2. **Cross-Validation**:  
   - For small datasets, perform **k-fold cross-validation**, which involves splitting the training data into several folds.  
   - Train the model on \( k-1 \) folds and validate it on the remaining fold, repeating this process \( k \) times.  
   - The model's performance is averaged across all validation folds to get a more reliable estimate of its generalization error.

**Considerations:**
- If the validation set is **too small**, model evaluation may be imprecise.  
- If the validation set is **too large**, the model may not train well due to insufficient training data.

---

### 9. Data Mismatch

If your training data does not represent the real-world data the model will encounter, this creates a **data mismatch** problem. Both the **validation set** and **test set** must be representative of the production data to ensure accurate evaluation.

**Challenges:**  
- Determining whether poor performance is due to bad data or an over/underperforming model can be difficult.

**Solution: Train-Dev Set (Proposed by Andrew Ng)**  
- **Train-Dev Set**: A subset of the training data that is not used during training but is representative of production data.  
  - If the model performs poorly on the train-dev set, it indicates **overfitting** to the training data.  
  - If the model performs well on the train-dev set but poorly on the validation set, the **training data is likely not representative** of production data.  

**Workflow:**
1. Train the model and evaluate on the **train-dev set**.  
2. If the model performs well on both the train-dev and validation sets, it is ready for testing on the test set.  
3. Once the model performs well on the test set, it can be scaled to production.
